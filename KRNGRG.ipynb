```python
import operator
fname = 'D:\\GoogleDrive\\A002_PRIVAT\\20190610_NathanAugust\\KRNGRG.txt'
right_name = 'NathanAugust'
right_date = '10.06'
penalty = 4

l1 = right_name.count('N')
l2 = right_name.count('a')
l3 = right_name.count('t')
l4 = right_name.count('h')
l5 = right_name.count('n')
l6 = right_name.count('A')
l7 = right_name.count('u')
l8 = right_name.count('g')
l9 = right_name.count('s')
l0 = right_name.count('t')

summe_right = l1+l2+l3+l4+l5+l6+l7+l8+l9+l0

with open(fname) as f:
    content = f.readlines()
content = [x.strip() for x in content] 

for i in range(len(content)):
    content[i] = content[i].replace(' ','')

content_dict = {}
for i in range(len(content)):
    content_key,content_value = content[i].split(sep=':')
    content_dict.update({content_key:content_value})

dict_punkte = {}
dict_anonym = {}
for i in content_dict:
    l1 = content_dict[i].count('N')
    l2 = content_dict[i].count('a')
    l3 = content_dict[i].count('t')
    l4 = content_dict[i].count('h')
    l5 = content_dict[i].count('n')
    l6 = content_dict[i].count('A')
    l7 = content_dict[i].count('u')
    l8 = content_dict[i].count('g')
    l9 = content_dict[i].count('s')
    l0 = content_dict[i].count('t')
    
    temp_summe = l1+l2+l3+l4+l5+l6+l7+l8+l9+l0
    
    summe = summe_right-temp_summe
    
    if content_dict[i].count('Nathan') == 1:
        bonus1 = 0
    else:
        bonus1 = penalty
        
    if content_dict[i].count('Natan') == 1:
        bonus2 = 0
    else:
        bonus2 = penalty
        
    if content_dict[i].count('August') == 1:
        bonus3 = 0
    else:
        bonus3 = penalty
    
    if content_dict[i].count(right_date)==1:
        bonus4 = 0
    else:
        bonus4 = penalty
        
    dict_punkte.update({i:summe+bonus1+bonus2+bonus3+bonus4})
    dict_anonym.update({content_dict[i]:summe+bonus1+bonus2+bonus3+bonus4})
    
sorted_punkte = sorted(dict_punkte.items(), key=operator.itemgetter(1))
sorted_anonym = sorted(dict_anonym.items(), key=operator.itemgetter(1))
sorted_anonym
```
